### **DATA IMPORTATION**

#### **Notes on Data Importation into Python**

##### **Introduction**

Data importation is the process of bringing data from external sources into Python for analysis, visualization, machine learning, or reporting. Python supports importing data from many formats such as CSV, Excel, databases, JSON, web APIs, and text files.

Python provides powerful libraries that simplify data importation and manipulation.

Commonly used libraries include:

* `pandas`
* `numpy`
* `sqlite3`
* `sqlalchemy`
* `mysql-connector-python`
* `openpyxl`
* `json`

---

#### **1. Importing CSV Files**

CSV (Comma Separated Values) files are among the most commonly used data formats.

#### **Using Pandas**

```python
import pandas as pd

df = pd.read_csv("data.csv")

print(df.head())
```

#### **Important Parameters**

| Parameter  | Purpose                     |
| ---------- | --------------------------- |
| `sep`      | Specifies delimiter         |
| `header`   | Row containing column names |
| `usecols`  | Select specific columns     |
| `nrows`    | Read limited rows           |
| `skiprows` | Skip rows                   |
| `encoding` | File encoding               |

#### **Example**

```python
df = pd.read_csv(
    "data.csv",
    sep=",",
    usecols=["Age", "Income"]
)
```

---

#### **3. Importing Text Files**

#### **Reading Simple Text File**

```python
with open("notes.txt", "r") as file:
    content = file.read()

print(content)
```

#### **Reading Line by Line**

```python
with open("notes.txt", "r") as file:
    for line in file:
        print(line)
```

---

#### **4. Importing JSON Data**

JSON is widely used in APIs and web applications.

#### **Reading JSON File**

```python
import pandas as pd

df = pd.read_json("data.json")
print(df.head())
```

#### **Using JSON Module**

```python
import json

with open("data.json") as file:
    data = json.load(file)

print(data)
```

#### **5. Importing Data from SQL Databases**

Python can connect to relational databases such as:

* MySQL
* PostgreSQL
* SQLite
* SQL Server

---

#### **A. Importing from MySQL**

#### **Install Required Packages**

```bash
pip install mysql-connector-python pandas
```

#### **Example**

```python
import mysql.connector
import pandas as pd

connection = mysql.connector.connect(
    host='localhost',
    port=3306,
    user='root',
    password='1992',
    database='mydb'
)

query = "SELECT * FROM orders"

df = pd.read_sql(query, connection)

print(df.head())

connection.close()
```

---

#### **B. Importing from SQLite**

SQLite requires no server installation.

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("database.db")

df = pd.read_sql("SELECT * FROM students", conn)

print(df.head())

conn.close()
```

#### **6. Importing Data from Web APIs**

APIs provide online data access.

#### **Install Requests Library**

```bash
pip install requests
```

#### **Example**

```python
import requests
import pandas as pd

url = "https://api.example.com/data"

response = requests.get(url)

data = response.json()

df = pd.DataFrame(data)

print(df.head())
```

#### **7. Importing Data from HTML Tables**

Python can scrape tables directly from websites.

```python
import pandas as pd

url = "https://example.com/table"

tables = pd.read_html(url)

df = tables[0]

print(df.head())
```

#### **8. Importing Large Datasets Efficiently**

Large datasets may consume significant memory.

#### **Reading in Chunks**

```python
chunks = pd.read_csv("largefile.csv", chunksize=1000)

for chunk in chunks:
    print(chunk.head())
```

#### **11. Best Practices in Data Importation**


#### **C. Handle Missing Values Early**

