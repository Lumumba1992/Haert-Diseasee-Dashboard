#### **1. Data Importation**  
Data importation is the first step in any data analysis workflow. In Python, the **pandas** library is widely used for loading data from various sources, such as CSV, Excel, SQL databases, JSON, and even web scraping. Proper data importation ensures that the dataset is structured correctly and is ready for further preprocessing.  

- **CSV Files**: The most common format for storing tabular data, where values are separated by commas.  
- **Excel Files**: Useful for structured data with multiple sheets.  
- **JSON and XML Files**: Common in web applications and APIs.  
- **SQL Databases**: Data stored in relational databases can be imported using SQL queries.  

When importing data, it is important to inspect the structure to confirm the correct delimiter, encoding, and column names. Errors in data importation, such as incorrect data types or encoding mismatches, may require additional preprocessing.  

#### **2. Data Cleaning**  
Real-world datasets often contain missing values, duplicates, and inconsistencies that must be addressed before analysis. Data cleaning ensures that the dataset is structured correctly and free from errors that might affect the accuracy of the results.  

##### **Handling Missing Data**  
Missing values can occur due to data entry errors, incomplete records, or system issues. There are several ways to handle missing data:  
- **Deletion**: Removing rows or columns with missing values if they are minimal.  
- **Imputation**: Replacing missing values with statistical measures such as the mean, median, or mode.  
- **Forward or Backward Fill**: Filling missing values using previous or next available values in time-series data.  

##### **Handling Outliers**  
Outliers are extreme values that differ significantly from the rest of the data and can distort analysis results. They can be detected using statistical methods such as:  
- **Z-score method**: Identifies data points that are several standard deviations away from the mean.  
- **Interquartile Range (IQR) method**: Identifies outliers based on quartiles.  
- **Visualization methods**: Box plots and scatter plots help in detecting extreme values.  

#### **3. Summary Statistics**  
Summary statistics provide insights into the distribution, central tendency, and spread of the data. Some of the key statistical measures include:  
- **Mean**: The average value of a dataset.  
- **Median**: The middle value when the data is sorted.  
- **Mode**: The most frequently occurring value.  
- **Variance and Standard Deviation**: Measures the spread of data around the mean.  
- **Skewness and Kurtosis**: Used to understand the shape of the distribution.  

These statistics help analysts understand the nature of the dataset and whether further transformations are necessary before modeling.  

#### **4. Exploratory Data Analysis (EDA)**  
Exploratory Data Analysis (EDA) is the process of analyzing datasets to summarize their key characteristics, often with the help of visualizations. It helps to identify patterns, trends, and potential issues within the data.  

##### **Data Visualization**  
Visualization is a key part of EDA, as it provides an intuitive understanding of data relationships. Common types of visualizations include:  
- **Histograms**: Show the distribution of numerical data.  
- **Box Plots**: Identify outliers and the spread of data.  
- **Scatter Plots**: Show relationships between two numerical variables.  
- **Bar Charts**: Compare categorical data.  
- **Heatmaps**: Display correlations between multiple numerical variables.  

Using **Matplotlib** and **Seaborn**, analysts can create these visualizations to better understand the data before applying statistical or machine learning models.  

#### **5. Importance of EDA**  
EDA is a critical step before building models, as it helps in:  
- Understanding the data structure and identifying inconsistencies.  
- Detecting missing values, outliers, and unusual patterns.  
- Selecting appropriate features for predictive modeling.  
- Improving data preprocessing and transformation steps.  

By the end of this session, learners will be able to import datasets, clean data by handling missing values and outliers, compute summary statistics, and create visualizations for exploratory data analysis.

##### **Fill missing values with median**

```python
import pandas as pd
import warnings
warnings.simplefilter("ignore")

# Specify the column to fill missing values (e.g., 'income')
column_name = "income"

# Replace missing values with the mean of the column
df[column_name].fillna(df[column_name].median(), inplace=True)

# Verify if missing values are replaced
print(f"Missing values in '{column_name}' after filling:", df[column_name].isnull().sum())
```

#### **REplace the Missing Value with the Mean for all Numeric Columns**

```python
import pandas as pd
# Select only numeric columns
numeric_cols = df.select_dtypes(include=['number']).columns

# Replace missing values in numeric columns with their mean
df[numeric_cols] = df[numeric_cols].apply(lambda col: col.fillna(col.mean()))

# Verify if missing values are replaced
print("Missing values after filling:\n", df[numeric_cols].isnull().sum())
```

#### **Identification of Outliers Using BoxPlots**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot box plots for numerical columns
plt.figure(figsize=(12, 6))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(y=df[col])
    plt.title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()
```

#### **Removing Outliers**

#### **1. 1.5 Inter-Quartile Rule**

```python
# Define the column to check for outliers (e.g., 'income')
column_name = "income"

# Calculate the Interquartile Range (IQR)
Q1 = df[column_name].quantile(0.25)
Q3 = df[column_name].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers
df = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]

print("Outliers removed successfully!")
```

#### **2.Capping (Winsorization)**
Instead of removing outliers, extreme values can be capped to a predefined threshold (e.g., replacing extreme values with the 5th and 95th percentiles).

```python
import numpy as np

# Define lower and upper caps
lower_cap = df["income"].quantile(0.05)  # 5th percentile
upper_cap = df["income"].quantile(0.95)  # 95th percentile

# Apply capping
df["income"] = np.where(df["income"] < lower_cap, lower_cap, df["income"])
df["income"] = np.where(df["income"] > upper_cap, upper_cap, df["income"])

print("Outliers have been capped!")
```

#### **Add the Line of Best Fit to the Scatter Plot**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import linregress

# Remove missing values to avoid 'nan' issues
df_clean = df[['age', 'income']].dropna()

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(df_clean['age'], df_clean['income'])

# Create scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df_clean['age'], y=df_clean['income'], color='blue', alpha=0.6, label="Data Points")

# Plot regression line manually
x_vals = np.linspace(df_clean['age'].min(), df_clean['age'].max(), 100)
y_vals = slope * x_vals + intercept
plt.plot(x_vals, y_vals, color='red', label="Best Fit Line", linewidth=2)

# Equation of the line
equation_text = f"y = {slope:.2f}x + {intercept:.2f}"

# Position equation inside the plot
plt.text(df_clean['age'].min() + 2, df_clean['income'].max() - 5000, equation_text, 
         fontsize=12, color='red', bbox=dict(facecolor='white', alpha=0.5))

# Add labels and title
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Scatter Plot of Age vs. Income with Best Fit Line")

# Show legend
plt.legend()

# Show plot
plt.show()
```
