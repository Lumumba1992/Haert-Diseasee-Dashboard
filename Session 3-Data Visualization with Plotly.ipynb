### **Scatter Plot of GDP Per Capita and Life Expectancy**

```python
# Load Gapminder dataset
df = px.data.gapminder()

# Create scatter plot
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df,
    x="gdpPercap",
    y="lifeExp"
)

# Add titles and labels
plt.title("GDP per Capita vs Life Expectancy")
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")

# Show plot
plt.show()
```

```python
plt.figure(figsize=(12, 7))

sns.scatterplot(
    data=df,
    x="gdpPercap",
    y="lifeExp",
    hue="continent",
    size="pop",
    alpha=0.7
)

plt.xscale("log")  # Better scaling for GDP
plt.title("GDP per Capita vs Life Expectancy by Continent")
plt.xlabel("GDP per Capita (log scale)")
plt.ylabel("Life Expectancy")

plt.show()
```

```python
# Load Gapminder dataset
df = px.data.gapminder()

# Create figure
plt.figure(figsize=(12, 7))

# Scatter plot
sns.scatterplot(
    data=df,
    x="gdpPercap",
    y="lifeExp",
    hue="continent",   # Color by continent
    size="pop",        # Bubble size by population
    alpha=0.7
)

# Improve x-axis scaling
plt.xscale("log")

# Titles and labels
plt.title(
    "Relationship Between GDP per Capita and Life Expectancy",
    fontsize=18,
    fontweight="bold"
)

plt.xlabel("GDP per Capita (Log Scale)", fontsize=14)
plt.ylabel("Life Expectancy", fontsize=14)

# Improve legend
plt.legend(
    bbox_to_anchor=(1.05, 1),
    loc="upper left"
)

# Add grid
plt.grid(True, linestyle="--", alpha=0.5)

# Tight layout
plt.tight_layout()

# Show plot
plt.show()
```

### **Interactive Scatter Plot**

```python
# Import libraries
import plotly.express as px

# Load Gapminder dataset
df = px.data.gapminder()

# Create interactive scatter plot
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    color="continent",
    size="pop",
    hover_name="country",
    log_x=True,
    size_max=60,
    title="GDP per Capita vs Life Expectancy by Continent",
    labels={
        "gdpPercap": "GDP per Capita (Log Scale)",
        "lifeExp": "Life Expectancy"
    }
)

# Improve layout
fig.update_layout(
    template="plotly_dark",   # Interactive dark theme
    title_x=0.5,              # Center title
    font=dict(size=14)
)

# Show interactive chart
fig.show()
```

```python
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    animation_frame="year",
    animation_group="country",
    size="pop",
    color="continent",
    hover_name="country",
    log_x=True,
    size_max=60,
    range_x=[100, 100000],
    range_y=[20, 90],
    title="Gapminder Animated Scatter Plot"
)

fig.show()
```

```python
# Install kaleido if not installed
# pip install kaleido imageio

import plotly.express as px
import imageio
import os

# Load dataset
df = px.data.gapminder()

# Create output folder
os.makedirs("frames", exist_ok=True)

# Generate frames for each year
years = sorted(df["year"].unique())

for year in years:
    
    # Filter data for a single year
    dff = df[df["year"] == year]
    
    # Create figure
    fig = px.scatter(
        dff,
        x="gdpPercap",
        y="lifeExp",
        size="pop",
        color="continent",
        hover_name="country",
        log_x=True,
        size_max=60,
        range_x=[100, 100000],
        range_y=[20, 90],
        title=f"Gapminder: {year}"
    )
    
    # Improve layout
    fig.update_layout(
        template="plotly_dark",
        title_x=0.5
    )
    
    # Save frame as PNG
    fig.write_image(f"frames/{year}.png", width=1000, height=600)

# Create GIF
images = []

for year in years:
    images.append(imageio.imread(f"frames/{year}.png"))

# Save GIF
imageio.mimsave(
    "gapminder_animation.gif",
    images,
    duration=1.2   # seconds per frame
)

print("GIF saved successfully!")
```

### **Interactive Histogram with Plotly**

```python
# Create interactive histogram
fig = px.histogram(
    df,
    x="lifeExp",
    nbins=30,
    hover_data=df.columns,
    title="Distribution of Life Expectancy"
)

# Improve layout
fig.update_layout(
    template="plotly_dark",
    title_x=0.5,
    bargap=0.05,
    font=dict(size=14)
)

# Show chart
fig.show()
```

#### **Annimate with Time**

```python
# Import libraries
import plotly.express as px

# Load Gapminder dataset
df = px.data.gapminder()

# Create animated interactive histogram
fig = px.histogram(
    df,
    x="lifeExp",
    color="continent",
    animation_frame="year",   # Animate over time
    nbins=30,
    opacity=0.8,
    barmode="overlay",
    title="Animated Distribution of Life Expectancy Over Time",
    labels={
        "lifeExp": "Life Expectancy",
        "count": "Frequency"
    }
)

# Slow down animation
#fig.layout.updatemenus[0].buttons[0].args[1]["frame"]["duration"] = 1200
#fig.layout.updatemenus[0].buttons[0].args[1]["transition"]["duration"] = 800

# Improve layout
fig.update_layout(
    template="plotly_dark",
    title_x=0.5,
    bargap=0.05,
    font=dict(size=14)
)

# Show chart
fig.show()
```
